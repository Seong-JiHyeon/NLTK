```
def stem(word):
  for suffix in ['ing', 'ly', 'ed', 'ious', 'ies', 'ive', 'es', 's', 'ment']:
    if word.endswith(suffix):
      return word[:-len(suffix)]
  return word
```

```
giRaws={}
for fileid in GNI.fileids():
  giRaws[fileid] = GNI.raw(fileid)
  giRaws[fileid] = re.sub(r'[\n\r.,()]', '', giRaws[fileid])
  words=nltk.word_tokenize(giRaws[fileid])
  words = [x for x in words if not (x.lower() in stopwords.words('english'))]
  fd = nltk.FreqDist(words)
  fd = sorted([(freq, word) for (word, freq) in fd.items()], reverse=True)
  print(fileid, fd[0:10])
```

* Regular Expression

`(?x)` : set flag to allow verbose regexps

`([A-Z]\.)+` : abbreviations (ex U.S.A)

`\w+(-\w+)*` : words with optional internal hyphens (ex hot-water) to one word

`\$?\d+(\.\d+)?%?` : currency percentages (ex $12.40, 82%)

`\.\.\.` : ...

[][.,;"'?():-_`] : seperate tokens include [ , ]

```
text = 'That U.S.A poster-print costs $12.40...'
pattern = r'''(?x)
    ([A-Z]\.)+
  | \w+(-\w+)*
  | \$?\d+(\.\d+)?%?
  | \.\.\.
  | [][.,;"'?():-_`]'''

nltk.regexp_tokenize(text, pattern)
```
