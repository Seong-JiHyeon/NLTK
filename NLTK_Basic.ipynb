```
from sklearn import datasets
import pandas as pd
iris = datasets.load_iris()

labels = pd.DataFrame(iris.target)
labels.columns=['labels']
data = pd.DataFrame(iris.data)
data.columns = ['Sepal length', 'Sepal width', 'Petal length', 'Petal width']
data = pd.concat([data, labels], axis=1)

data.head()
```

```
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

feature = data[['Sepal length', 'Sepal width']]
model = KMeans(n_clusters=3, algorithm='auto')
model.fit(feature)
predict = pd.DataFrame(model.predict(feature))
predict.columns=['predict']

r = pd.concat([feature, predict], axis=1)
```

```
centers = pd.DataFrame(model.cluster_centers_, columns=['Sepal length', 'Sepal width'])
center_x = centers['Sepal length']
center_y = centers['Sepal width']

plt.scatter(r['Sepal length'], r['Sepal width'], c=r['predict'], alpha =0.5)
plt.scatter(center_x, center_y, s=50, marker='D', c='r')
plt.show()
```

**학습활동**: myBigram, myCollocation함수 정의


```
myList = ['time', 'files', 'like', 'an', 'arrow'] #{}표시는 집합을 뜻함
def myBigram(myList):
  result = []
  for i in range(len(myList)-1):
    result.append((myList[i],myList[i+1]))
  return result
```

```
myBigram(text4)
```

```
def myCollocation(myList):
  bigram_words = myBigram(myList)
  bigram_freqs = FreqDist(bigram_words)
  return bigram_freqs.most_common(50)
```

```
myCollocation(text4)
```

**Automatic Natural Language**

* 자연언어처리단계:

> 1. 형태소(morphology) 분석: 친구였다 = 친구 + 이 + 었 + 다
>2. 통사(syntax) 분석: 그가 사과를 먹는다 = 주어 + 목적어 + 동사
> 3. 의미(semantics) 분석: plant = industrial, living organism
> 4. 화용(pragmatics) 분석: 네 뒤에 벼랑이 있는지 알고 있니? = 위험하다

* word sense disambiguation: identifying which sense of word(meaning) is used in sentence => semantic role labeling

>Ex. bass = fish, guitar, tones of low frequency

* pronoun resolution (anaphora resolution)

:identifying what a pronoun or noun phrase refers to (semantic role labeling>)

> 1. The thieves stole the paintings. **They**(paintings) were subsequently sold.
> 2. The thieves stole the paintings. **They**(thieves) were subsequently caught.
> 3. The thieves stole the paintings. **They**(both) were subsequently found.

* subtasks that information extraction system can perform include:

>1. chunking(term analysis)
>2. named-entity recognition: date, time, name, organisation, unit, percentage
>3. fact extraction

* Machine translation

=> using probability in machine translation: 사람이 직접 규칙이나 예외를 학습시키는 것이 아닌, 실제 번역된 수많은 사례들의 학습을 통해 스스로 패턴을 찾게 해서 학습시킴. 사례를 많이 학습 할수록 성능은 좋아짐

=> '초파리가 바나나를 좋아한다'를 구글 번역은 '과일은 바나나처럼 날아간다'로 번역함. '시간은 화살처럼 날아간다'와 같이 학습된 사례가 많기 때문에 번역에 오류가 발생
